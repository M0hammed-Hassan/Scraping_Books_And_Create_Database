```python
#Library
import sqlite3
```

```python
#Connect to database
connection=sqlite3.connect('Books.db')
cursor=connection.cursor()
```

```python
#Write a sql query to get books that has rate >=3 and has 'Mr' in its name.
cursor.execute('''SELECT * FROM Books 
                  WHERE Rate>=? AND Title LIKE ?''',[3,'%Mr%'])
rows=cursor.fetchall()
for name,categ,rate,price in rows:
    print(f"{name}.............{categ}.............{rate}.............{price}")
```

```python
#Write a sql query to get books that has rate >=3 or price > 20.
cursor.execute('''SELECT * FROM Books 
                  WHERE Rate>=? OR Price > ?''',[3,20])
rows=cursor.fetchall()
for name,categ,rate,price in rows:
    print(f"{name}.............{categ}.............{rate}.............{price}")
```

```python
#Write a sql query to get books that has rate not 3.
cursor.execute('''SELECT * FROM Books
                  WHERE Rate != ?''',[3])
rows=cursor.fetchall()
for name,categ,rate,price in rows:
    print(f"{name}.............{categ}.............{rate}.............{price}")
```

```python
#Write a sql query to get books that has price between 10 and 40 and has rate 3 or 4 or 1.
cursor.execute('''SELECT * FROM Books
                  WHERE Price BETWEEN ? AND ? AND Rate IN(?,?,?)''',[10,40,3,4,1])
rows=cursor.fetchall()
for name,categ,rate,price in rows:
    print(f"{name}.............{categ}.............{rate}.............{price}")
```

```python
#Write a sql query to get the top 5 most expensive books (order books desc with price column and then limit the result to the first 5).
cursor.execute('''SELECT * FROM Books
                  ORDER BY Price DESC LIMIT 5 ''')
rows=cursor.fetchall()
for name,categ,rate,price in rows:
    print(f"{name}.............{categ}.............{rate}.............{price}")
```

```python
#Write a sql query to get the 3rd 10 books in the books table order first by rate desc and then by price asc.
cursor.execute('''SELECT * FROM Books
                  ORDER BY Rate DESC,Price ASC LIMIT 10 OFFSET 3''')
rows=cursor.fetchall()
for name,categ,rate,price in rows:
    print(f"{name}.............{categ}.............{rate}.............{price}")
```

```python
#Write a sql query to add a new category in categories table and add 5 books to this category in books table with title, rate and price.
cursor.execute('''INSERT INTO Categories(Name) VALUES (?)''',['Engineering'])
connection.commit()
En_categ=['Math1','Math2','Physics','Python Programming','C Programming']
for i in range(5):
    cursor.execute('''INSERT INTO Books(Title,Category_Name,Rate,Price) VALUES(?,?,?,?)''',[En_categ[i],'Engineering',i+1,i+40])
    connection.commit()
```

```python
#Write a sql query to update book's rate to 3 that thier price < 20£.
cursor.execute('''UPDATE  Books SET Rate=3 WHERE Price<?''',[20])
connection.commit()
```

```python
#Write a sql query to delete all books that have price > 50£ and has rate <= 2.
cursor.execute('''DELETE FROM Books WHERE Price> ? AND Rate<=?''',[50,2])
connection.commit()
```

```python
#Write a sql query to count the number of books that have 'Secret' in thier names and price between 10£ and 25£.
cursor.execute('''SELECT COUNT(Title) FROM Books WHERE Title LIKE ? AND Price BETWEEN ? AND ?''',['%Secret%',10,25])
row=cursor.fetchall()
print(row[0][0])
```

```python
#Write a sql query to get the minimum & maximum price for all the books that have rate 5.
row=cursor.execute('SELECT MIN(Price),MAX(Price) FROM Books WHERE Rate=?',[5]).fetchall()
print(row[0][0],' ',row[0][1])
```

```python
#Write a sql query to calculate the avg price for all the books that have rate 5.
cursor.execute('''SELECT AVG(Price) FROM Books WHERE Rate=5''')
row=cursor.fetchall()
row[0][0]
```

```python
#Write a sql query to sum all book's price that have rate 2 and price between 10 and 40
row=cursor.execute('''SELECT SUM(Price) FROM Books WHERE Rate=2 AND Price BETWEEN 10 AND 40''').fetchall()
row[0][0]
```

```python
#Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.
#Acutally i don't need to join here as my database contains in table Books foreign key Category_Name from table Categories
row=cursor.execute('''CREATE TABLE Details AS
                        SELECT Title AS Book_Name,
                        Category_Name AS Book_Category,
                        Rate AS Book_rate,
                        Price AS Book_Price
                        FROM Books''')
                        
connection.commit()
```

```python
#Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.
row=cursor.execute('''SELECT COUNT(Price) FROM Books WHERE Price BETWEEN 20 AND 30''').fetchall()
row[0][0]
```

```python
#Write a sql query to calculate how many books each category has having count > 10.
row=cursor.execute('''SELECT Category_Name ,COUNT(*) AS Cnt FROM Books GROUP BY Category_Name HAVING cnt>10''').fetchall()
len(row)
```
