```python
#Library
import pandas as pd
import sqlite3
```

```python
#Connect to database
connection=sqlite3.connect('Books.db')
cursor=connection.cursor()
```

```python
#Display table Books
df=pd.read_sql('''SELECT * FROM Books''',connection)
df
```

```python
#Export table Books to csv
df.to_csv('Books_Table.csv',index=False)
```
